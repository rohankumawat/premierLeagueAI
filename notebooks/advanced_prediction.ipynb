```python
matches.dtypes
```

```python
del matches["comp"]
```

```python
del matches["notes"]
```

```python
# change the date datatype from object to date
matches["date"] = pd.to_datetime(matches["date"])
```

```python
matches.dtypes
```

```python
# Create a multi-class target column: 0 = Loss, 1 = Draw, 2 = Win
matches["target"] = matches["result"].apply(lambda x: 2 if x == "W" else (1 if x == "D" else 0))
```

```python
matches["target"].value_counts()
```

```python
# Convert categorical columns to numerical codes
matches["venue_code"] = matches["venue"].astype("category").cat.codes
matches["opp_code"] = matches["opponent"].astype("category").cat.codes
```

```python
# Convert 'time' column to hour
matches["hour"] = matches["time"].str.replace(":.+", "", regex=True).astype("int")
```

```python
# Extract day of the week from the 'date' column
matches["day_code"] = matches["date"].dt.dayofweek
```
