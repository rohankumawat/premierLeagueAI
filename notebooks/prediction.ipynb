## Improving precision with Rolling Averages

```python
# split the matches data frame by team bcoz what we want to do is compute 
# "rolling averages" for team performance and how did this team do in the past few games?
#Â how many shots did they have?
# how many goals were scored against them?, etc. etc.
grouped_matches = matches.groupby("team")
```

```python
group = grouped_matches.get_group("Manchester City")
```

```python
group.head()
```

```python
# if we are at matchweek 4, how did manchester city performed in the past 3 matchweeks!
def rolling_averages(group, cols, new_cols):
    group = group.sort_values("date") # ascending order (look at the last 3 matches)
    rolling_stats = group[cols].rolling(3, closed='left').mean() # closed = 'left' takes the current week out
    group[new_cols] = rolling_stats
    group = group.dropna(subset=new_cols)
    return group
```

```python
cols = ["gf", "ga", "sh", "sot", "dist", "fk", "pk", "pkatt"]
new_cols = [f"{c}_rolling" for c in cols] # add rolling on to the end of the column names
new_cols
```

```python
# just for the manchester city games
rolling_averages(group, cols, new_cols)
```

```python
matches_rolling = matches.groupby("team").apply(lambda x: rolling_averages(x, cols, new_cols))
```

## Retraining our Machine Learning Model

```python
def make_predictions(data, predictors):
    # split the data into training and test set
    train = data[data["date"] < '2022-01-01']
    test = data[data["date"] > '2022-01-01']
    # random forest
    rf.fit(train[predictors], train["target"])
    preds = rf.predict(test[predictors])
    # combine the predictions and actuals together
    combined = pd.DataFrame(dict(actual=test["target"], predicted=preds), index=test.index)
    # calculate precision
    error = precision_score(test["target"], preds)
    return combined, error
```

```python
combined, error = make_predictions(matches_rolling, predictors + new_cols)
```

```python
# fix the Wolverhamptom Wanderers and Wolves issue

class MissingDict(dict):
    __missing__ = lambda self, key: key

map_values = {"Brighton and Hove Albion": "Brighton", 
              "Manchester United": "Manchester Utd", 
              "Newcastle United": "Newcastle Utd", 
              "Tottenham Hotspur": "Tottenham", 
              "West Ham United": "West Ham", 
              "Wolverhampton Wanderers": "Wolves"}

mapping = MissingDict(**map_values) # replace it with the original names
```

```python
combined["new_team"] = combined["team"].map(mapping)
```

```python
merged = combined.merge(combined, left_on=["date", "new_team"], right_on=["date", "opponent"])
```
